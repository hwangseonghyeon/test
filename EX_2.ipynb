```python
digits = load_digits()
print(digits.data.shape)
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(digits.images[0])
plt.show()
```

```python
import pandas as pd
digits_data = digits.data
digits_label = digits.target

print(digits.target)
digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names)
digits_df
```

```python
digits_df["target"] = digits.target
digits_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=1)
# 여러번 구동해도 값이 바뀌지 않게 하기 위해 1로 설정했습니다


print('X_train 개수: ', len(x_train),', X_test 개수: ', len(x_test))
```

```python
# 노드에서 원하는 모델을 만들기 위해 필요한 라이브러리 전부 미리 import

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
```

```python
decision_tree = DecisionTreeClassifier(random_state=1)
# random state로 인해 값이 매번 바뀌는 현상을 없애기 위해 1로 설정했습니다
decision_tree.fit(x_train, y_train)

y_pred = decision_tree.predict(x_test)

# print(y_pred)
# print(y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

# 데이터가 중구난방해서 fake pred는 만들어서 인공지능 성능을 확인할 필요가 없다
# 판단하여 fake_pred는 만들지 않았습니다.
```

```python
random_forest = RandomForestClassifier(random_state=1)
random_forest.fit(x_train, y_train)
y_pred = random_forest.predict(x_test)

# print(y_pred)
# print(y_test)

print(classification_report(y_test, y_pred))
```

```python
svm_model = svm.SVC()

svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)

# print(y_pred)
# print(y_test)

print(classification_report(y_test, y_pred))
```

```python
sgd_model = SGDClassifier()

sgd_model.fit(x_train, y_train)
y_pred = sgd_model.predict(x_test)

print(classification_report(y_test, y_pred))
```

```python
logistic_model = LogisticRegression()

# 해당 코드를 실행하면, 아직 정답에 수렴하지 않았다는 문구가 뜨는데,
# 이는 max_iter값을 늘려줘야 되지만,
# 늘려주면 stop:total no. of iterations reched limit 라는 문구가 뜬다
# 이는 1을 넣어도 뜨고 기본값인 100을 넣어도 뜨는데
# 해결방안은 아직 찾지 못했다
# 찾아본 결과 정답에 수렴하지 않았을 뿐 모델이 만들어지긴 했다고 한다
# LogisticRegression(max_iter=100)


logistic_model.fit(x_train, y_train)
y_pred = logistic_model.predict(x_test)

print(classification_report(y_test, y_pred))
```

classification_report로 모델의 성능을 볼 때,
가장 정확도가 높고, 값들이 안정적인 svm_model 모델이기 때문에 svm_model을 선택하고자 합니다
